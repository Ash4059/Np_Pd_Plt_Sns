Matplotlib

```python
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("Line plot")
plt.xlabel(
    xlabel="X-axis"
)
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=2)
plt.title("Line plot")
plt.xlabel(
    xlabel="X-axis"
)
plt.ylabel("Y-axis")
plt.show()
```

```python
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,color='red',linestyle=":",linewidth=5)
plt.title("Line plot")
plt.xlabel(
    xlabel="X-axis"
)
plt.grid(True)
plt.ylabel("Y-axis")
plt.show()
```

plt.subplot(rows,column,row num,col num)

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.subplot(1,2,2)
plt.plot(x,y2,color='red',linestyle=":",linewidth=5)
plt.show()
```

Bar plot

```python
student={"Bob":87,"Matt":56,"Sam":27}
name=list(student.keys())
marks=list(student.values())
```

```python
plt.bar(name,marks)
plt.show()
```

```python
plt.bar(name,marks)
plt.title("Bar plot")
plt.xlabel('Names')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

Horizontal bar plot

```python
plt.barh(name,marks,color='red')
plt.title("Bar plot")
plt.xlabel('marks')
plt.ylabel('names')
plt.grid(True)
plt.show()
```
